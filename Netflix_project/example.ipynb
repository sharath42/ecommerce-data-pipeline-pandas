```python
df2=pd.read_csv("C:/Users/91728/OneDrive/Desktop/Pandas_prac/Netflix_project/salaries_10k.csv")
df2.info
```

```python
df=df1.merge(df2, left_on='emp_id', right_on='emp_id')
df
```

```python
df['total_salary']=df['salary']+df['bonus']
df_filter=df[(df['salary']>3000)& (df['department']!='Sales')]
df_filter=df_filter.groupby(['join_date','total_salary','bonus']).agg({'salary':'sum','department':'count'})
df_filter=df_filter.rename(columns={'total_salary':'final_salary'},inplace=False)
df_filter
```
