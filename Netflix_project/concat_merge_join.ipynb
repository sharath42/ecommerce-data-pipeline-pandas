```python
import pandas as pd

df1=pd.read_csv('employees_10k.csv')

df1.info()
```

```python
df2=pd.read_csv('salaries_10k.csv')
df2.info()
```

```python
df1=df1.drop_duplicates()
df2=df2.drop_duplicates()
df3=df1.merge(df2,left_on='emp_id',right_on='emp_id')
df4=df3.dropna()
df4['total_salary']=df4['salary']+df4['bonus']
df_filter=df4[df4['salary']<100000]
df_filter=df_filter.rename(columns={'department':'department_name'},inplace=False)
df_final=df_filter.groupby(['join_date','department_name','emp_name','total_salary']).agg({'salary':'sum',
                                                                                        'bonus':'mean'})
df_final=df_final.sort_values(by=['department_name','join_date','salary'],ascending=False)
df_final
#df_final.to_csv("C:/Users/91728/OneDrive/Desktop/Pandas_prac/Netflix_project/final_output.csv")
```
