```python
import pandas as pd

d={'categories':['A','B','C','A'],'Values1':[20,30,40,50],'Values2':[50,60,70,60],'items':['sha','yas','meh','gou']}
df=pd.DataFrame(d)
df
```

```python
#multiple cloumns isused for agg in group by
test1=df.groupby(['categories','items']).agg({'Values1':'count','Values2':'sum'})  


test1
```

```python
# single coulmn is used for group by
t=df.groupby(['categories','items'])['Values1'].sum()
t
```
